### Define a function to get data (Using Discog API client )

```python
def get_all_user_ids(username):
    user_dict = {}
    artist_ids = []
    release_ids = []
    user = discogsclient.user(username)
    #wantlist
    try:
        for page in range(user.wantlist.pages):
            time.sleep(0.1)
            page_releases_list = user.wantlist.page(page)
            for rel in page_releases_list:
                time.sleep(0.1)
                artist_id = rel.data['basic_information']['artists'][0]['id']
                release_id = rel.id
                release_ids.append(release_id)
                artist_ids.append(artist_id)
    finally:
        pass
    #collection
    try:
        for release in user.collection_folders[0].releases:
            time.sleep(0.1)
            col_rel_id = release.id
            release_ids.append(col_rel_id)
            col_art_id = release.data['basic_information']['artists'][0]['id']
            artist_ids.append(col_art_id)
    finally:
        pass
    user_dict['user'] = username
    user_dict['artist_ids'] = artist_ids
    user_dict['release_ids'] = release_ids
    return user_dict
```
